```python
import spotipy
from spotipy.oauth2 import SpotifyOAuth

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(
    client_id = client_ID,
    client_secret= client_Secret,
    scope="user-top-read",
    redirect_uri = 'http://localhost:8000'
    ))
```

```python
import numpy as np
import pandas as pd

column = ['name','artist','spotify_id','danceability','energy','loudness','acousticness','instrumentalness','valence','tempo']
top_songs = pd.DataFrame(columns = column)


results = sp.current_user_top_tracks(limit=50, offset=0, time_range='short_term')
tracks = results['items']

def get_features(track_id):
    #grabs features from songs. There are more. Check the Spotify API Audio Feature object
    analysis = sp.audio_features(tracks= track_id)
    danceability = analysis[0]['danceability']
    energy = analysis[0]['energy']
    loudness = analysis[0]['loudness']
    acousticness = analysis[0]['acousticness']
    instrumentalness = analysis[0]['instrumentalness']
    valence = analysis[0]['valence'] 
    tempo = analysis[0]['tempo'] 

    return danceability,energy,loudness,acousticness,instrumentalness,valence,tempo

for i in range(len(tracks)):
    name= tracks[i]['name']
    artist= tracks[i]['artists'][0]['name']
    danceability,energy,loudness,acousticness,instrumentalness,valence,tempo = get_features(track_id)

    top_songs = top_songs.append({'name':name,
                                'artist':artist,
                                'spotify_id':track_id,
                                'danceability':danceability,
                                'energy':energy,
                                'loudness':loudness,
                                'acousticness':acousticness,
                                'instrumentalness':instrumentalness,
                                'valence':valence,
                                'tempo':tempo},
                                ignore_index = True)
    

print(top_songs.sort_values('tempo'))



#grab top artists
# results = sp.current_user_top_artists(limit=5, offset=0, time_range='short_term')
# artists = results['items']

# # for i in range(len(artists)):
# #     print(artists[i]['name'])
```
